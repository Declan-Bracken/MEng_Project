```python
"""
I need a heuristic algorithm here to determine whether the page is double or single column, and if splitting is necessary.
"""
from PIL import Image
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
import cv2

# filename = '/content/Screen Shot 2024-03-22 at 4.24.53 PM.png'
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image
filename = '/content/transcripts-2-2048.png'
img = cv2.imread(filename, 0)  # Read the image as grayscale

# Define the percentage of the image to crop from the top and bottom
crop_percent = 15  # Adjust as needed

# Calculate the pixel indices for cropping
height, width = img.shape
crop_top = int(height * crop_percent / 100)
crop_bottom = height - crop_top

# Crop the image
cropped_img = img[crop_top:crop_bottom, :]

# Binarization using Otsu's thresholding
_, binary_img = cv2.threshold(cropped_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Plot the image
plt.figure(dpi = 300)  # You can adjust the figure size as needed
plt.imshow(binary_img)
plt.axis('off')  # This hides the axis ticks and labels
plt.show()

# Whitespace Distribution Analysis
# Compute histogram of white pixels along the vertical axis
histogram = np.sum(binary_img == 255, axis=0)

# Plotting the histogram
plt.figure(figsize=(10, 5))
plt.plot(histogram, color='black')
plt.title('Vertical Whitespace Distribution')
plt.xlabel('Pixel Column')
plt.ylabel('White Pixel Count')
plt.grid(True)
plt.show()

max_whitespace = cropped_img.shape[0] # calculate max pixels
avg_histo = np.mean(histogram)
center_val = (max_whitespace - histogram[len(histogram)//2-1])/max_whitespace # fractional area of whitespace in the center vertical line
avg_val = (max_whitespace - avg_histo)/ max_whitespace # average fractional area of non-whitespace everywhere

threshold_factor = 5
# Determine if there is a significant increase in the white pixel count in the center valiue compared to average
if avg_val > threshold_factor * center_val:
    print("DOUBLE COLUMN LAYOUT")
else:
    print("SINGLE COLUMN LAYOUT")

```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_name_or_path = "TheBloke/CapybaraHermes-2.5-Mistral-7B-GPTQ"
# To use a different branch, change revision
# For example: revision="gptq-4bit-32g-actorder_True"
model = AutoModelForCausalLM.from_pretrained(model_name_or_path,
                                             device_map="auto",
                                             trust_remote_code=False,
                                             revision="gptq-4bit-32g-actorder_True")

tokenizer = AutoTokenizer.from_pretrained(model_name_or_path, use_fast=True)
```

```python
# Function to get a response from mistral given an OCR text string
def query_mistral(text, do_sample = False, temperature=0.3, top_p=0.9, top_k=40, repetition_penalty=1.1):
    # prompt = f'''Below is OCR text from a student transcript. This text contains a table, or multiple tables. I want you to select only the course codes, course names, and course grades from this text, then organize these 3 fields into a csv file.

    # ### Text:
    # {text}
    # '''
    prompt = f'''Below is OCR text from a student transcript. This text contains a table, or multiple tables. I want you to select data relevant to student courses and grades from these tables and format the fields into a table in csv format.

    ### Text:
    {text}
    '''

    system_message = "You are a table creation assistant"
    prompt_template=f'''<|im_start|>system
    {system_message}<|im_end|>
    <|im_start|>user
    {prompt}<|im_end|>
    <|im_start|>assistant
    '''
    if do_sample == False:
      temperature=None
      top_p=None
      top_k=None

    # print("*** Pipeline:")
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=2048,
        do_sample= do_sample,
        temperature=temperature,
        top_p=top_p,
        top_k=top_k,
        repetition_penalty=repetition_penalty
    )
    LLM_response = pipe(prompt_template)[0]['generated_text']
    return LLM_response
```

```python
def extract_after_keyword(text, keyword = "<|im_start|>assistant"):
    """
    Extracts and returns the part of the text that comes after the first occurrence of the specified keyword.
    If the keyword is not found, the function returns an empty string.
    """
    # Find the position of the keyword in the text
    keyword_pos = text.find(keyword)

    # If the keyword is found, extract the part after the keyword
    if keyword_pos != -1:
        return text[keyword_pos + len(keyword):].strip()

    # If the keyword is not found, return an empty string
    print(f"Keyword '{keyword}' not found in text:\n{text}")
    return ""
```

```python
import re

def extract_csv_content(text):
    # Define the regex pattern for matching CSV rows with two comma separations
    # This pattern looks for lines with any non-comma characters, separated by commas
    csv_pattern = re.compile(r'^[^,]+,[^,]+,[^,\n]+$', re.MULTILINE)

    # Search for all matches in the text
    matches = csv_pattern.findall(text)

    # Join the matched lines into a single string to form the CSV content
    csv_content = '\n'.join(matches)

    return csv_content

def extract_csv_content_flexible(text):
    # Split the entire text into parts separated by at least one empty line
    parts = re.split(r'\n\s*\n', text, maxsplit=1)

    # Assume the second part (if exists) contains the CSV data
    if len(parts) > 1:
        csv_part = parts[1]
    else:
        # If there's no clear separation, consider the whole text
        csv_part = parts[0]

    # Capture lines that contain at least one comma
    csv_lines = re.findall(r'^[^,]+,[^,\n]+(?:,[^,\n]+)*$', csv_part, re.MULTILINE)

    # Join the captured lines to form the CSV content
    csv_content = '\n'.join(csv_lines)

    return csv_content


def extract_csv_content_after_empty_line(text):
    # Split the entire text into parts separated by at least one empty line
    parts = re.split(r'\n\s*\n', text, maxsplit=1)

    # Check if we have a part that potentially contains the CSV data
    if len(parts) > 1:
        csv_part = parts[1]
    else:
        # If there's no empty line, assume the whole text is the part to consider
        csv_part = parts[0]

    # Now, capture lines with exactly two comma separations within the CSV part
    csv_lines = re.findall(r'^[^,]+,[^,]+,[^,\n]+$', csv_part, re.MULTILINE)

    # Join the captured lines to form the CSV content
    csv_content = '\n'.join(csv_lines)

    return csv_content

# Example usage
# extracted_text = """
# Here is the requested CSV file containing the course codes, course names, and course grades:

# Course Code, Course Title, Course Grade
# ART 114, Basic Drawing, A
# ENGL 248, Intro to Literature, B-
# IPS 100, First Year Seminar, B+
# IDS 200, Encounter with Cultures, A
# KIN 140, Health Dynamics, A-
# ART 105, Basic Design, A-
# ART 111, Intro to Art History, A-
# EDUC 220, Educational Psychology, Bt
# EDUC 221, Educ Psych Field Placement, A
# ENGL 113, The Creative Impulse, Bt
# REL 100, From Rabbi to God, A-
# ART 113, Basic Painting, Bt
# ART 117, Basic Ceramics, A-
# EDUC 225, The Exceptional Child, Bt
# EDUC 226, Exceptnl Child Field Placement, Bt
# POL 151, Intro to Global Politics, B+
# SPAN 121, Spanish I, EXM
# SPAN 195, Intensive Beginning Spanish, B-
# """

# csv_content = extract_csv_content_after_empty_line(extracted_text)
# print(csv_content)
```

```python
import pandas as pd
from collections import Counter
from io import StringIO

def adjust_row_fields(row, target_fields):
    """
    Adjust a row to have a specific number of fields, either by merging excess fields
    or padding with None for missing fields.
    """
    if len(row) > target_fields:
        # Merge excess fields; this strategy may need customization
        row = row[:target_fields-1] + ['; '.join(row[target_fields-1:])]
    elif len(row) < target_fields:
        # Pad row with None values
        row += [None] * (target_fields - len(row))
    return row

def preprocess_csv_data(csv_data):
    # Determine the most common number of fields in rows
    rows = [row.split(',') for row in csv_data.split('\n') if row]
    field_counts = Counter(len(row) for row in rows)
    common_field_count = field_counts.most_common(1)[0][0]

    # Adjust rows to have uniform number of fields
    adjusted_rows = [adjust_row_fields(row, common_field_count) for row in rows]

    # Convert adjusted rows to CSV string
    adjusted_csv_str = '\n'.join(','.join(str(field) for field in row) for row in adjusted_rows)

    # Load into DataFrame
    df = pd.read_csv(StringIO(adjusted_csv_str))

    return df

# # Example CSV data with varying number of fields
# csv_data = """2018 Fall/Winter, Course Code, Course Name, Credits, Grade
# w,CALCULUS 1000B,CALCULUS I,0.5,0.67
# w,ICOMPSCI 1032A,INFORMATION SYSTEMS & DESIGN,0.5,0.72
# w,FAECONOMIC 1021A,PRINCIPLES OF MICROECONOMICS,0.5,0.76
# w,RaECONOMIC 1022B,PRINCIPLES OF MACROECONOMICS,0.5,0.72
# w,[SIJAPANESE 1036,JAPANESE FOR BEGINNERS,as OG, OFA,1.78"""

# # Preprocess and load the CSV data into a DataFrame
# df = preprocess_csv_data(csv_data)
# print(df)

```

```python
class AnomalyDetection:
    def __init__(self, df):
        self.df = df

    def find_missing_values(self):
        missing_values = self.df.isnull()
        missing_cells = missing_values.stack()
        missing_indices = missing_cells[missing_cells].index.tolist()
        return missing_indices

    def check_data_ranges(self):
        numeric_columns = self.df.select_dtypes(include=['number']).columns
        data_ranges = {}
        for col in numeric_columns:
            if len(self.df[col].unique()) > 1:  # Skip if only one unique value
                data_ranges[col] = (self.df[col].min(), self.df[col].max())
        return data_ranges

    def unique_value_count(self):
        unique_counts = {}
        for col in self.df.columns:
            unique_counts[col] = self.df[col].nunique()
        return unique_counts

# Example usage:
# Assuming df is your DataFrame
# anomaly_detector = AnomalyDetection(df)
# missing_rows, missing_columns = anomaly_detector.find_missing_values()
# data_ranges = anomaly_detector.check_data_ranges()
# unique_counts = anomaly_detector.unique_value_count()
```
