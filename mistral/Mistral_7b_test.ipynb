```python
prompt_template =f'''You are my table creation assistant. Below is text from a transcript with a table, or multiple tables. I want you to select only the course codes, course names, and course grades from this text, then organize these 3 fields into a csv file. Ensure you do not miss any courses.

### Text:
{OCR_text}

### CSV Table:
'''
```

```python
import torch

# Tokenize the text and count the number of tokens
tokens = tokenizer.tokenize(prompt_template)
num_tokens = len(tokens)
print(f"Number of tokens: {num_tokens}")
# Enable mixed-precision to optimize GPU usage
torch.cuda.amp.autocast(enabled=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def generate_code(prompt, max_length=2048):
    # Encode the prompt to the format the model expects
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    # Generate a response
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1)

    # Decode the generated tokens to a string
    generated_code = tokenizer.decode(output[0], skip_special_tokens=True)

    return generated_code

generated_code = generate_code(prompt_template)

print("\nINITIAL GENERATED CSV:\n")
print(generated_code)
```

```python
# To improve robustness, I'll add a self-verification step to the model:
# Split the generated code to extract only the CSV part
csv_part_generated = generated_code.split("### CSV Table:")[-1].strip()

# Adding a verification step
verification_prompt = f'''You have generated a CSV table. Please review the original text, compare it with the generated CSV table, and add any additional courses that may have been missed into an updated table.
### Generated CSV Table:
{csv_part_generated}

### Orginial text:
{OCR_text}

### Updated CSV Table:
'''

verified_output = generate_code(verification_prompt)
print("Verified and corrected CSV if necessary:")
print(verified_output)
```
